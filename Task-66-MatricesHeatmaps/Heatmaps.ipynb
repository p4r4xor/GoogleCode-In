<font size="4">More fun!</font>

```python
from astropy.modeling import models
```

```python
z = np.zeros((100, 100))
y, x = np.mgrid[0:100, 0:100]
```

```python
for _ in range(10):
    gauss = models.Gaussian2D(amplitude=100, x_mean=np.random.randint(0, 100), y_mean=np.random.randint(0, 100), x_stddev=3, y_stddev=3)
    z += gauss(x, y)
    airy = models.AiryDisk2D(amplitude=70, x_0=np.random.randint(0, 100), y_0=np.random.randint(0, 100), radius=5)
    z += airy(x, y)
```

```python
plt.figure()
```

```python
plt.plot(z)
```

```python
plt.contour(z)
```

```python
plt.contour(z, cmap='ocean')
```
